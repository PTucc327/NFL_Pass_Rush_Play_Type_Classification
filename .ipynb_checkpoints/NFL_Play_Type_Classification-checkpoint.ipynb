## Introduction

In the NFL, there are 32 teams each choosing plays of the same type(pass,run). During the 2024 season passing plays were 58%, and running plays were 42%. While this is just the average of the league as a whole, each team has a slightly different percentage of pass and run plays. The plays teams choose to run varies based on game situations(ex. 3rd and 1 vs 3rd and 10). Understanding these choices is important in uncovering team strategies, and their decision making process.

## Goal

To create a Machine Learning model that can predict what type of play will occur based on game situation.

## Feature Engineering

```python
df= df.drop('yards_gained', axis=1)
df_copy = df[df['play_type_nfl'].isin(['PASS', 'RUSH'])].copy()
df_copy['play_type_nfl'] = df['play_type_nfl'].map({'PASS': 1, 'RUSH': 0})

```

```python
features = [
    'down', 'ydstogo', 'yardline_100', 'qtr', 'drive',
    'half_seconds_remaining', 'score_differential',
    'posteam', 'defteam', 'home_team','away_team', 
    'shotgun', 'away_wp','home_wp', 'no_huddle','qb_dropback', 
    'total_home_rush_epa', 'total_away_rush_epa', 'total_home_pass_epa','total_away_pass_epa',
    'home_timeouts_remaining','away_timeouts_remaining', 
    'defense_coverage_type', 'offense_formation'
]
```

```python
categorical_features=['home_team','away_team','posteam','defteam','qb_dropback', 'shotgun', 'no_huddle','defense_coverage_type', 'offense_formation']
numeric_features = [ 'down', 'ydstogo', 'yardline_100', 'qtr', 'drive',
    'half_seconds_remaining', 'score_differential','away_wp','home_wp','total_home_rush_epa', 'total_away_rush_epa', 'total_home_pass_epa','total_away_pass_epa',
    'home_timeouts_remaining','away_timeouts_remaining']
```

```python
from sklearn.model_selection import train_test_split
X= df_copy.drop('play_type_nfl', axis=1).copy()
y= df_copy['play_type_nfl']
```

```python
from sklearn.preprocessing import LabelEncoder

# Encode y once
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y_encoded, test_size = 0.2, random_state= 42, stratify= y)
```

## Modeling

### Imports

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
import xgboost as xgb
```

### Logistic Regression

```python
ohe = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
scaler = StandardScaler()

preprocessor = ColumnTransformer(
    transformers=[
        ('cat', ohe, categorical_features),
        ('num', scaler, numeric_features)
    ]
)

# Pipeline
log_reg_model = Pipeline(steps=[
    ("prep", preprocessor),
    ("model", LogisticRegression(max_iter=500))
])

# Fit the FULL pipeline (this also fits the preprocessor!)
log_reg_model.fit(X_train, y_train)

# Evaluate
y_pred = log_reg_model.predict(X_test)
```

```python
log_reg_model.fit(X_train,y_train)
```

```python
y_pred_lr = log_reg_model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
```

```python
print(f'Accuracy of the model is:  {accuracy}')
print(f'Classification Report of the model is: \n {report}')
```

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
cm_display = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=log_reg_model.classes_)
cm_display.plot()
plt.title("Confusion Matrix")
plt.show()
```

### RandomForest Classifier

```python
from sklearn.pipeline import Pipeline
rf = Pipeline([
    ('prep', preprocessor),
    ('model', RandomForestClassifier())
])

```

```python
rf.fit(X_train,y_train)
```

```python
y_pred_rf = rf.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred_rf)
report = classification_report(y_test, y_pred_rf)
```

```python
print(f'Accuracy of the model is:  {accuracy}')
print(f'Classification Report of the model is: \n {report}')
```

```python
cm = confusion_matrix(y_test, y_pred_rf)
```

```python
cm_display = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=rf.classes_)
cm_display.plot()
plt.title("Confusion Matrix")
plt.show()
```

### XGBoost

```python
from xgboost import XGBClassifier


xgb_model = XGBClassifier(
    n_estimators=300,
    max_depth=6,
    learning_rate=0.1,
    subsample=0.8,
    colsample_bytree=0.8,
    eval_metric='logloss'
)

pipeline_xgb = Pipeline(steps=[
    ("prep", preprocessor),   # SAME preprocessor you use for every model
    ("model", xgb_model)
])


```

```python
pipeline_xgb.fit(X_train, y_train)
```

```python
y_pred_xgb = pipeline_xgb.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred_xgb)
report = classification_report(y_test, y_pred_xgb)
```

```python
print(f'Accuracy of the model is:  {accuracy}')
print(f'Classification Report of the model is: \n {report}')
```

```python
cm = confusion_matrix(y_test, y_pred_xgb)
```

```python
cm_display = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=pipeline_xgb.classes_)
cm_display.plot()
plt.title("Confusion Matrix")
plt.show()
```

## Evaluating Model performance

```python
def predict_play_type(model,
                      down=None,
                      ydstogo=None,
                      yardline_100=None,
                      qtr=None,
                      drive=None,
                      half_seconds_remaining=None,
                      score_differential=None,
                      posteam=None,
                      defteam=None,
                      home_team=None,
                      away_team=None,
                      shotgun=None,
                      away_wp=None,
                      home_wp=None,
                      no_huddle=None,
                      qb_dropback=None,
                      total_home_rush_epa=None,
                      total_away_rush_epa=None,
                      total_home_pass_epa=None,
                      total_away_pass_epa=None,
                      home_timeouts_remaining=None,
                      away_timeouts_remaining=None,
                      defense_coverage_type=None,
                      offense_formation=None):

    # Raw 1-row dataframe
    sample = {
        'down': down,
        'ydstogo': ydstogo,
        'yardline_100': yardline_100,
        'qtr': qtr,
        'drive': drive,
        'half_seconds_remaining': half_seconds_remaining,
        'score_differential': score_differential,
        'posteam': posteam,
        'defteam': defteam,
        'home_team': home_team,
        'away_team': away_team,
        'shotgun': shotgun,
        'away_wp': away_wp,
        'home_wp': home_wp,
        'no_huddle': no_huddle,
        'qb_dropback': qb_dropback,
        'total_home_rush_epa': total_home_rush_epa,
        'total_away_rush_epa': total_away_rush_epa,
        'total_home_pass_epa': total_home_pass_epa,
        'total_away_pass_epa': total_away_pass_epa,
        'home_timeouts_remaining': home_timeouts_remaining,
        'away_timeouts_remaining': away_timeouts_remaining,
        'defense_coverage_type': defense_coverage_type,
        'offense_formation': offense_formation
    }

    X_sample = pd.DataFrame([sample])

    # Send raw sample directly into the pipeline
    pred = model.predict(X_sample)[0]

    return "PASS" if pred == 1 else "RUN"
```

```python
predict_play_type(
    pipeline_xgb,
    down=3,
    ydstogo=8,
    yardline_100=50,
    qtr=4,
    drive=12,
    half_seconds_remaining=180,   # 3 minutes left
    score_differential=-7,        # losing by 7
    posteam="KC",
    defteam="BUF",
    home_team="KC",
    away_team="BUF",
    shotgun=1,
    away_wp=0.65,
    home_wp=0.35,
    no_huddle=0,
    qb_dropback=1,
    total_home_rush_epa=12.3,
    total_away_rush_epa=8.4,
    total_home_pass_epa=25.1,
    total_away_pass_epa=18.7,
    home_timeouts_remaining=2,
    away_timeouts_remaining=3,
    defense_coverage_type="COVER_2",
    offense_formation="SHOTGUN"
)
```

# Results

Overall, this project gave insights into team strategies as well as if machine learning can predict play calls. While looking at the data, it is clear teams have different amount of call types but overall similar structure with passing being the most common play type followed by rushing plays. It is also clear that teams have different averages for the amount of yards gained based on play types which can be due various factors. Some of the factors include players on the field, and how the offensive play calls and defensive formation calls matter.

When looking at the modeling section, all models performed very well. Evaluating the results generated, the XGBoost model was the best performing model not only through metrics but also through the confusion matrix showing similar results for classifying rushing plays but had the best results for classifying passing plays. In the evaluation process, I performed 2 types of testing, first being self generated data to test whether the model can predict between pass and rush, which did not prove the value of the model which then led me to my second option. I used the same API as before to collect data from 2025, using this current data, I ran the data through the model and visualized and evaluated the performance. As seen with the modeling section the performance was up to same standards as shown with my expirement. This means my model is also able to generalize really well, which can be due to using pre play information and nothing specific like players. It also is more current data which helps with current team patterns. 

The results of the models also shows something more meaningful. If play calls can be easily predicted, games would be very challening as teams wouldn't be able to progress down field, as the defense would know what the offense will do. Teams have to use trick plays or run different types of plays in different scenarios to try and throw the defense off of their game and progress down the field to score. This can even be noticed in games, for example, when watching the New York Giant's, rookie quarterback Jaxson Dart can be seen throwing short passes throughout the game, but when the game gets closer to the end and he is looking to score he will throw a deep pass even if the scenario would suggest doing something different. It can even be noticed in games where a running play is called and the quarterback decides to keep it and pass it off to a receiver or the complete opposite they are looking to throw and a gap opens up and they run with the ball.

For future adaptations, this project can be utilized in a few ways. Starting off, it can be experimented with players on field, and see how players impact play call decisions. Second, this can turn into an application that can be used for teams to help decide play calls during games or during practices to prepare for upcoming games.
